#### Импорт библиотек и инициализация переменных <a class="anchor" id="1"></a>

```python
import numpy as np
```

```python
m_rows = 85
m_cols = 85

n_rows = 85
n_cols = 85

p_size = 85
q_size = 85

mean = 2
dispersion = 4
std = np.sqrt(dispersion)
```

#### Формирование матриц $\mathbf{M}$ и $\mathbf{N}$ <a class="anchor" id="2"></a>

Сформируем матрицу $\mathbf{M}$

```python
m = np.full(
    shape=(m_rows, m_cols),
    fill_value=2,
    dtype=int
)

np.fill_diagonal(m, np.arange(3, 172, 2))

m
```

Сформируем матрицу $\mathbf{N}$

```python
n = np.full(
    shape=(n_rows, n_cols),
    fill_value=4,
    dtype=int
)

n[::2,2::3] = 5
n[1::2,1::2] = 7

n
```

#### Формирование векторов $\mathbf{p}$ и $\mathbf{q}^T$ <a class="anchor" id="3"></a>

Сформируем вектор-столбец $\mathbf{p}$

```python
p = np.array([np.random.normal(mean, std, p_size) + 1]).T
p
```

Сформируем вектор-строку $\mathbf{q}^T$

```python
q = np.array([n[24] * n[71]])
q
```

#### Вычисление значения функции $\mathbf{f}$ <a class="anchor" id="4"></a>

```python
q_square = q ** 2
q_square
```

```python
q_square_argmax = q_square.argmax()
q_square_argmax
```

```python
f = 4 * n[q_square_argmax, 59]
f
```

#### Вычисление значения функции $\mathbf{g}$ <a class="anchor" id="5"></a>

```python
m_dot_p = m.dot(p)
print(m_dot_p.shape)
print(m_dot_p)
```

```python
m_dot_p_mul_q = m_dot_p * q
print(m_dot_p_mul_q.shape)
print(m_dot_p_mul_q)
```

```python
m_dot_p_mul_q_mul_constant = m_dot_p_mul_q * 7e-6
m_dot_p_mul_q_mul_constant
```

```python
m_dot_p_mul_q_mul_constant_square = m_dot_p_mul_q_mul_constant ** 2
m_dot_p_mul_q_mul_constant_square
```

```python
submatrix = m_dot_p_mul_q_mul_constant_square[24:70,13:28]
print(submatrix.shape)
print(submatrix)
```

```python
submatrix_row_elements_mul = submatrix.prod(axis=1)
print(submatrix_row_elements_mul.shape)
print(submatrix_row_elements_mul)
```

```python
np.sum(submatrix_row_elements_mul)
```
